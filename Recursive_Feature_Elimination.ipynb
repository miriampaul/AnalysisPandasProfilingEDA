```python
import pandas as pd
```

```python
f=pd.read_csv("/content/bank.csv")
```

```python
#implement Recursive Feature Elimination
#Predict wther the customer is going to buy the product or not
#Long duration of call , better
#Drop the duration column
f=f.drop("duration", axis=1)
```

```python
#SPlit the columns into Dependent(Y) and Independent(X)
x=f.iloc[:, :-1]
y=f.iloc[:, -1]
```

```python
#Create dummmy variables
x=pd.get_dummies(x, drop_first=True)
y=pd.get_dummies(y, drop_first=True)
```

```python
#Split the dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=\
train_test_split(x, y, test_size=0.3, random_state=1234)
```

```python
#Import randomforest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
rfc=RandomForestClassifier(random_state=1234)
rfc.fit(X_train, Y_train)
Y_predict=rfc.predict(X_test)
#Evaluation the model
cm2=confusion_matrix(Y_test, Y_predict)
score2=rfc.score(X_test, Y_test)
```
