Adult Income Prediction Problem Understanding:
Predict whether income exceeds 50K per year based on census data:
Adult income datasets : Age, wc, education, marital status, race, gender, hours per week, income class



```python
import pandas as pd
data=pd.read_csv("/content/04+-+decisiontreeAdultIncome.csv")
```

```python
data1=data.copy()
```

```python
#Check for null values
data.isnull().sum(axis=0)
```

```python
#Check the datatypes
data.dtypes
```

```python
#Create dummy variables
data_prep=pd.get_dummies(data, drop_first=True)
```

```python
data_prep
```

```python
#Create the X and Y variables
X=data_prep.iloc[: , :-1]
```

```python
Y=data_prep.iloc[:, -1]
```

```python
Y
```

```python
#split the X and Y datasets into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=\
train_test_split(X, Y, test_size=0.3, random_state=1234, stratify=Y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Train the model
dtc=DecisionTreeClassifier(random_state=1234)
dtc.fit(X_train, Y_train)
Y_predict=dtc.predict(X_test)
```

```python
#Evaluate the model
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test, Y_predict)
score=dtc.score(X_test, Y_test)
```

```python
cm
```

```python
score
```

```python
#Import Random forest classifier from sklearn
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=1234)
rfc.fit(X_train, Y_train)
Y_predict=rfc.predict(X_test)
#Evaluate the model
cm2=confusion_matrix(Y_test, Y_predict)
score2=rfc.score(X_test, Y_test)
```
